```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import librosa
import librosa.display


audio_data = 'Data/genres_original/classical/classical.00096.wav'
data , sr = librosa.load(audio_data)
print(type(data), type(sr))

plt.figure(figsize=(12, 4))
librosa.display.waveplot(data, color = "#502A75")
plt.show()
```

```python
genres='blues classical country rock disco reggae hiphop jazz metal pop'.split()

os.chdir('Data/genres_original/')

for i,genre in enumerate(genres):
    y,sr=librosa.load(genre+'/'+genre+'.00005.wav',duration=5.0,sr=None,)
    librosa.display.specshow(librosa.feature.chroma_stft(y, sr=sr), x_axis='time', y_axis='chroma', cmap='coolwarm')
    plt.title(genre.upper()+"Chroma feature")
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
    plt.savefig(fname=genre+'_Chroma_feature',quality=95,dpi=200)

```

```python
#printing signal wave
genres='blues classical country rock disco reggae hiphop jazz metal pop'.split()
for i,genre in enumerate(genres):
    y,sr=librosa.load(genre+'/'+genre+'.00005.wav',duration=30.0,sr=None,)
    time=np.linspace(0,len(y)/sr,num=len(y))
    plt.figure(i+1)
    plt.figure(figsize=(15,5))
    plt.title(genre+" Signal Wave")
    plt.xlabel('Time')
    plt.yticks(ticks=[])
    plt.plot(time,y)
    #plt.savefig(fname=genre+'_time_domain',quality=95,dpi=200)
    plt.show()
```

```python
#printing Spectogram
for i,genre in enumerate(genres):
    y,sr=librosa.load(genre+'/'+genre+'.00005.wav',duration=5.0,sr=None,)
    librosa.display.specshow(librosa.amplitude_to_db(abs(librosa.stft(y)),ref=1.0),
                         y_axis='log', x_axis='time')
    plt.title(genre.upper()+" Power Spectrogram")
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
    #plt.savefig(fname=genre+'_power_spectrogram',quality=95,dpi=200)
    plt.show()
```

```python
##prtinting Spectrum

n_fft = 2048 # FFT window size
hop_length = 512 # number audio of frames between STFT columns

# Short-time Fourier transform (STFT)
spec = np.abs(librosa.stft(data, n_fft = n_fft, hop_length = hop_length))

#print('Shape of spec object:', np.shape(spec))

plt.figure(figsize = (16, 6))
plt.title('Spectrum', fontsize=25)
plt.xlabel('Frequency Bin', fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.plot(spec)
plt.savefig('spectrum.png')
```

```python
#printing Spectogram
for i,genre in enumerate(genres):
    y,sr=librosa.load(genre+'/'+genre+'.00005.wav',duration=5.0,sr=None,)
    librosa.display.specshow(librosa.amplitude_to_db(abs(librosa.stft(y)),ref=1.0),
                         y_axis='log', x_axis='time')
    plt.title(genre.upper()+" Power Spectrogram")
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
    plt.savefig(fname=genre+'_power_spectrogram',quality=95,dpi=200)
    plt.show()
```
